## 1. Teorema de Fermat (an치lisis)

Si una funci칩n $f(x)$ alcanza un m치ximo o m칤nimo local en $x=c$, y si la derivada $f'(c)$ existe en el punto $c$, entonces $f'(c) = 0$.

### Ejemplo

Sabemos que la funci칩n $f(x)=(x-2)^2$ tiene un m칤nimo global en $x=2$, pues

$$f(x)=(x-2)^2\geq0,\qquad\text{y}\qquad f(x)=(x-2)^2=0 \qquad\text{si y solo si}\qquad x=2.$$

### Pasos a seguir
1. Derivamos
  1. Definimos la variable simb칩lica
  2. Definimos la funci칩n a derivar
  3. Ahora s칤 derivamos 游때
2. Resolvemos para la derivada igualada a cero
3. Evaluamos si se trata de un m치ximo o m칤nimo
  1. Graficamos
  2. Evaluamos en puntos definidos
  3. Criterio de la segunda derivada



```python
# 1. Derivamos.
#    1. Definimos la variable simb칩lica
sym.var('x')
```

```python
#    1.2. Definimos la funci칩n a derivar
f=(x-2)**2
f # Mostramos la variable simb칩lica 'f'
```

```python
#     1.3. Ahora s칤 derivamos
df=sym.diff(f,x) # Derivamos 'f' con respecto a 'x'
df
```

```python
# 2. Resolvemos para la derivada igualada a cero
c=sym.solve(df)
c # Esta es una lista de raices
```

### En varias variables...

El procedimiento es an치logo.

Si una funci칩n $f:\mathbb{R}^n\to\mathbb{R}$ alcanza un m치ximo o m칤nimo local en $\boldsymbol{x}=\boldsymbol{c}\in\mathbb{R}^n$, y $f$ es diferenciable en el punto $\boldsymbol{x}=\boldsymbol{c}$, entonces $\left.\frac{\partial f}{\partial \boldsymbol{x}}\right|_{\boldsymbol{x}=\boldsymbol{c}}=\boldsymbol{0}$ (todas las derivadas parciales en el punto $\boldsymbol{x}=\boldsymbol{c}$ son cero).

**Criterio de la segunda derivada:** para ver si es m치ximo o m칤nimo, se toma la segunda derivada (matriz Hessiana) y se verifica definici칩n negativa o positiva, respectivamente.

Si se restringe a cierta regi칩n, hay ciertas t칠cnicas. La m치s general, pero tambi칠n la m치s compleja es la de **multiplicadores de Lagrange**.

**Ejemplo:** Determinar los valores extremo de $z(x,y)= (x-3)^2 + (y+5)^2$ usando el criterio de la segunda derivada y corroborarlo con la gr치fica.
